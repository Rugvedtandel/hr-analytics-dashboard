Average Salary

Department wise Attrition

```python
dept_attrition = df[df['Attrition']=='Yes'].groupby('Department').size()
print(dept_attrition)
```

Age Group Anlysis

```python
age_attrition = df[df['Attrition']=='Yes'].groupby('AgeGroup').size()
print(age_attrition)
```

Education Field Attrition

```python
education_attrition = df[df['Attrition']=='Yes'].groupby('EducationField').size().sort_values(ascending=False)
print(education_attrition)
```

Overtime Impact

```python
overtime_attrition = df[df['Attrition']=='Yes'].groupby('OverTime').size()
print(overtime_attrition)
```

Job satisfaction Anlysis

```python
job_sat = (df[df['Attrition']=='Yes'].groupby('JobSatisfaction').size())
print(job_sat)
```

Salary Slab Analysis

```python
salary_attrition = (df[df['Attrition']=='Yes'].groupby('SalarySlab').size())
print(salary_attrition)
```
